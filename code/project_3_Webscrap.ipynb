```python
import requests
import time
import pandas as pd
```

```python
posts = []
after = None
url = 'https://www.reddit.com/r/UpliftingNews.json'
for i in range(60):
    print(i)
    if after == None:
        param = {}
    else:
        param = {'after':after}
    url = 'https://www.reddit.com/r/UpliftingNews.json'
    res = requests.get(url,params = param, headers = {'User-agent':'abc0872*A'})
    if res.status_code == 200:
        the_json = res.json()
        list_of_posts = the_json['data']['children']
        data_posts = [post['data'] for post in list_of_posts]
        posts.extend(data_posts)
        after = the_json['data']['after']
        if after == None:
            continue
        else:
            url = 'https://www.reddit.com/r/UpliftingNews.json?after=' + after
        
    else:
        print(res.status_code)
        break
    time.sleep(2)
    
pd.DataFrame(posts).to_csv('../datasets/upnews.csv',index=False)    
```
